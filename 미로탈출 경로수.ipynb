```python
f = open("d:\kimsy\미로경로수in1.txt")
matrix = [list(map(int,f.readline().split())) for _ in range(7)]
dx=[-1,0,1,0]
dy=[0,1,0,-1]

def DFS(x,y): #Depth First Search(DFS)
    global cnt
    if x == 6 and y == 6:
        cnt +=1
    else:
        for i in range(4):
            xx = x + dx[i]
            yy = y + dy[i]
            if 0<= xx <=6 and 0 <= yy <= 6 and matrix[xx][yy] == 0:
                matrix[xx][yy] = 8
                beautiful(matrix)
                DFS(xx,yy)
                matrix[xx][yy] = 0
                beautiful(matrix)
def beautiful(matrix):
    for xx in matrix:
        print(xx)
    print()
        
cnt=0
matrix[0][0]=8
DFS(0,0)
print(cnt)
```
