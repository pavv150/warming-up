```python
from collections import deque
f = open("d:/kimsy/사과나무in1.txt")
dx=[-1,0,1,0]
dy=[0,1,0,-1]
n = int(f.readline())
matrix = [list(map(int, f.readline().split())) for _ in range(n)]
#초기화 작업
level,total = 0,0
Q = deque()
ck_tbl=[[0]*n for _ in range(n)]
total += matrix[n//2][n//2]
ck_tbl[n//2][n//2] = 1
Q.append((n//2,n//2))

while True:
    if level == n//2:
        break
    size = len(Q)
    for i in range(size):
        temp = Q.popleft()
        for j in range(4):
            x= temp[0] + dx[j]
            y= temp[1] + dy[j]
            if ck_tbl[x][y] == 0:
                total += matrix[x][y]
                ck_tbl[x][y] = 1
                Q.append((x,y))
                print("level={},size={}".format(level,size))
                for xx in ck_tbl:
                    print(xx)
                print()
    level += 1
print(total)
```
